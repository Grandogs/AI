```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import datetime
```

```python
# Dicionário de faturamento
dict_faturamento = {
    'data_ref': [
        '2023-01-01', 
        '2020-02-01', 
        '2021-03-01', 
        '2022-04-01', 
        '2023-05-01',
        '2023-06-01', 
        '2020-07-01', 
        '2021-08-01', 
        '2022-09-01', 
        '2023-10-01',
        '2022-11-01', 
        '2023-12-01',
        ],
    'valor': [
        400000, 
        890000, 
        760000, 
        430000, 
        920000,
        340000, 
        800000, 
        500000, 
        200000, 
        900000,
        570000, 
        995000,
        ]
}
```

```python
dict_faturamento
```

```python
df_faturamento = pd.DataFrame.from_dict(dict_faturamento)
```

```python
df_faturamento
```

```python
df_faturamento.info()
```

```python
date_col = pd.DatetimeIndex(df_faturamento['data_ref'])
```

```python
df_faturamento['mes'] = date_col.month
```

```python
df_faturamento['ano'] = date_col.year
```

### Média

```python
#média
int(df_faturamento["valor"].mean())
```

### Gráfico de Barra Vertical - Vendas

```python
df_faturamento.plot.bar(x='mes', y='valor')
```

```python
df_faturamento.sort_values('valor', ascending=False).plot.bar(x='ano', y='valor')
```

### Gráfico de Linhas - Vendas

```python
df_faturamento.plot.line(x='data_ref', y='valor')
```

```python
df_faturamento.sort_values('ano', ascending=True).plot.line(x='data_ref', y='valor')
```
